### while

````python
i = 0
while i < 10:
    i += 1
    print(i)
````

````python
# break continue     
# break   continue  
#   for   
i = 0
while i < 10:
    i += 1
    if i > 5:
        break
    print(i)
````

````python
i = 0
while i < 10:
    i += 1
    if i == 5:
        continue
    print(i)
````
